```python
solved_times = {
    "REINFORCE_advantage": 997,
    "A2C": 272,
    "A2C_tn": 312,
    "PPO_eps_02": 91,
    "PPO_eps_07": 83,
    "PPO_eps_1": 167
}
```

```python
from matplotlib import pyplot as plt

# Order the dictionary by values (solved times) in descending order
ordered_solved_times = dict(sorted(solved_times.items(), key=lambda x: x[1], reverse=True))

# Extract the keys and values
algorithms = list(ordered_solved_times.keys())
times = list(ordered_solved_times.values())

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(algorithms, times, color='skyblue')

# Add title and labels
plt.title('Environment Solved Times for Different Algorithms')
plt.xlabel('Algorithms')
plt.ylabel('Time (seconds)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```
